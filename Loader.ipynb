```python
mot = 0
y = 1
def motionsetter(x):
    global y
    global mot
    if (x == 0) & (y != 0):
        mot = mot + 1
    y = x
    return mot

df_mach2['StepNumber'] = df_mach2['Ground Speed (mph)'].apply(lambda x:motionsetter(x))
```

```python
df_mach2.tail(10)
```

```python
df_out = df_mach2.groupby(['StepNumber']).agg({'Ground Speed (mph)':'mean','OriginationTimeStamp':['max','min']})
```

```python
df_out['diff'] = (df_out['OriginationTimeStamp']['max'] - df_out['OriginationTimeStamp']['min'])/ np.timedelta64(1, 's')
df_out['dist'] = df_out['Ground Speed (mph)']['mean'] * df_out['diff'] / 3600
```

```python
#df_out.head(10) 
df_out['start'] = df_out['OriginationTimeStamp']['min']
df_out[df_out['dist']>0.04]
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
df_out[(df_out['OriginationTimeStamp']['max'] < pd.Timestamp('2016-11-14 20:00:00')) & (df_out['OriginationTimeStamp']['max'] > pd.Timestamp('2016-11-13 20:20:00')) & (df_out['dist']<1)].plot(ax=ax, kind='line', x='start', y='dist')
df[(df['StartDateTime'] < pd.Timestamp('2016-11-14 20:00:00')) & (df['StartDateTime'] > pd.Timestamp('2016-11-13 20:20:00'))].plot(ax=ax, kind='line', x='StartDateTime', y='Tonnage', secondary_y=True)
```
